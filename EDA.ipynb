```python
city_count = training_df['city'].value_counts()
print(city_count)
```

```python
plt.figure(figsize=(12, 6))
bars = sns.barplot(x=city_count.index, y=city_count.values, palette='Blues_d')
plt.title('Case Count by City')
plt.xlabel('City')
plt.ylabel('Case Count')
plt.xticks(rotation=0, ha='right')

for i, bar in enumerate(bars.patches):
    value = bar.get_height()
    bars.text(
        bar.get_x() + bar.get_width()/2., 
        value + 5,  
        f'{int(value)}', 
        ha='center',
        va='bottom',
        fontsize=11,
        fontweight='bold'
    )

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

cities = ['sj', 'iq']

for i, city in enumerate(cities):
    city_data = training_df[training_df['city'] == city]
    
    axes[i].scatter(x='weekofyear', y='total_cases', data=city_data)
    
    weekly_avg = city_data.groupby('weekofyear')['total_cases'].mean().reset_index()
    
    axes[i].plot(weekly_avg['weekofyear'], weekly_avg['total_cases'],
            color='red', linewidth=2, label='Weekly average')
    
    axes[i].set_title(f'Total Cases Week of Year - {city.upper()}', fontsize=14)
    axes[i].set_xlabel('Week of Year')
    axes[i].set_ylabel('Total Cases')
    axes[i].grid(True, linestyle='--', alpha=0.7)
    axes[i].legend()

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

cities = ['sj', 'iq']

for i, city in enumerate(cities):
    city_data = training_df[training_df['city'] == city]
    
    axes[i].scatter(x='year', y='total_cases', data=city_data)
    
    yearly_avg = city_data.groupby('year')['total_cases'].mean().reset_index()
    
    axes[i].plot(yearly_avg['year'], yearly_avg['total_cases'],
            color='red', linewidth=2, label='Yearly average')
    
    # Force integer tick labels on x-axis
    axes[i].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'{int(x)}'))
    
    # Ensure ticks are placed at integer positions
    min_year = city_data['year'].min()
    max_year = city_data['year'].max()
    axes[i].set_xticks(range(int(min_year), int(max_year)+1, 2))
    
    axes[i].set_title(f'Total Cases per Year - {city.upper()}', fontsize=14)
    axes[i].set_xlabel('Year')
    axes[i].set_ylabel('Total Cases')
    axes[i].grid(True, linestyle='--', alpha=0.7)
    axes[i].legend()

plt.tight_layout()
plt.show()
```

# Assessing Visual of Case Data:

Cases by Week of Year:
- strong seasonality in SJ: you can see there's a clear seasonal pattern with peak cases occurring around weeks 30-45 (roughly July-November)
- patter differs in IQ: there is a seasonal patter but it's not as visible as SJ, there are multiple small peaks throughout the year
- the difference in scale favors SJ. SJ peaks at 450+ cases compared to IQ that had roughly under 120 at it's peak

Cases per Year by City:
- massive outbreak in SJ around 1994, spike in cases occurred in this period
- there are years with much higher case counts than others in both cities
- different peak years between both cities
- SJ has 28yrs worth of data while IQ has 10

All Cases by Year:
- 1994 confirmed a major outbreak in SJ
- pattern shows that dengue doesn't simple increase over time but has cycles

# EDA Takeaways: 

1. Explore city specific modeling - patterns are significantly different between the two cities, might be worth building multiple models for each city. (feature engineering)
2. Create seasonal indicators based on week of year. (feature engineering)
3. Include lagged features (cases from previous weeks). (time series model)
4. Create features that would capture the time since the last outbreak. (feature engineering)
5. Use a time-based validation strategy rather than cross-validation (random selection) to account for the temporal nature of the outbreaks. (validation process for model)
6. Consider log-transforming the target variable (case count). (feature engineering)
7. Seek to reduce dimensionality by selecting one variable from each of the highly correlated cluster to avoid multicollinearity issues.
